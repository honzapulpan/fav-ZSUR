```python
import numpy as np
import matplotlib.pyplot as plt
#from tools import dist_m

import matplotlib.colors as pltcolors

plt.rcParams['figure.figsize'] = [10, 10]

data = np.genfromtxt('data.txt', delimiter=' ')
data_len = len(data)

plt.scatter(data[:,0],data[:,1], color='navy', marker='x')
plt.grid(True, which='both')
plt.axhline(y=0, color='gray')
plt.axvline(x=0, color='gray')

#vpoet potu td
from maximin import maximin
x = np.zeros(100)
mm = lambda t: maximin(data)
vfunc = np.vectorize(mm)
x = vfunc(x)
cl_nb = int(np.ceil(np.mean(x)))

# rozdlen dat do td
from kmeans import kmeans_direct, iterative_optim, kmeans_binary, plot_kmeans_result
#from tools import plot_result

eps = 1.e-03 #nastaven zastavovac podmnky

# pm dlen
data_classified = kmeans_binary(data, cl_nb, eps)
data_classified = iterative_optim(data_classified, cl_nb)
plot_kmeans_result(data_classified, 8)
```

```python
# pidn 1 na zatek vech dat
w_data = []
for i in range(cl_nb):
    ones = np.ones((len(data_classified[i]),1))
    w_data.append(np.concatenate((ones,data_classified[i]),axis=1))
#print(w_data)
```

```python
# vytvoen przdn matice prvk diskriminanch funkc f
fi = []
for i in range(1): #n
    fi.append([])
    for c in range(1): 
        fi[i].append(np.array([1,1,1]))
print(fi)
```

```python
for i in range(1):
    for j in range(1):
        min_len=np.amin([len(w_data[j+1]),len(w_data[i])])
        for n in range(min_len):
            if np.dot(fi[i][j],w_data[j+1][n,:].T)*(-1) < 0:
                fi[i][j] = fi[i][j] + (-1)*w_data[j+1][n,:]
            if np.dot(fi[i][j],w_data[i][n,:].T) < 0:
                fi[i][j] = fi[i][j] + w_data[i][n,:] 
                
print(fi)
```
